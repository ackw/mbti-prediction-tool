# Transformers & Transfer Learning

```python
df = data

labels=df['label'].unique().tolist()
IE_labels=df['IE'].unique().tolist()
NS_labels=df['NS'].unique().tolist()
TF_labels=df['TF'].unique().tolist()
JP_labels=df['JP'].unique().tolist()

df['label_int']=df['label'].apply(lambda x:labels.index(x))
df['IE_int']=df['IE'].apply(lambda x:IE_labels.index(x))
df['NS_int']=df['NS'].apply(lambda x:NS_labels.index(x))
df['TF_int']=df['TF'].apply(lambda x:TF_labels.index(x))
df['JP_int']=df['JP'].apply(lambda x:JP_labels.index(x))

df.head()
```

```python
# train, test, validate
train_val, test = train_test_split(df, test_size=0.1, random_state=42, shuffle=True)
train, val = train_test_split(train_val, test_size=0.2, random_state=42, shuffle=True)
train, val, test = map(lambda x:x.reset_index(drop=True), [train, val, test])
```

```python
import torch

class format_data(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)
```

## RoBERTa

### RoBERTa - All

Tokenize

```python
labels=df['label'].unique().tolist() # converts into list for processing
```

```python
# load tokenizer
from transformers import RobertaTokenizer
tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
tokenizer.max_model_input_sizes
```

```python
tokenized_len=pd.DataFrame({'length': list(map(len, tokenizer(df['text'].to_list(), truncation=False, padding=False)['input_ids']))})
tokenized_len.describe()
```

```python
ax = sns.histplot(tokenized_len)
ax.set(xlabel='Length of tokenized posts', ylabel='Count', xlim=(0, 1200), title='Distribution of the length of tokenized posts')
plt.show()
```

Fine-tuning

```python
# Format data
train_encodings = tokenizer(train['text'].to_list(), truncation=True, padding=True)
val_encodings = tokenizer(val['text'].to_list(), truncation=True, padding=True)
test_encodings = tokenizer(test['text'].to_list(), truncation=True, padding=True)

train_dataset = format_data(train_encodings, train['label_int'].to_list())
val_dataset = format_data(val_encodings, val['label_int'].to_list())
test_dataset = format_data(test_encodings, test['label_int'].to_list())
```

```python
# Fine-tuning
from transformers import RobertaForSequenceClassification, Trainer, TrainingArguments
model = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=len(labels)).to('cuda')
```

```python
training_args = TrainingArguments(output_dir='./results', num_train_epochs=3, per_device_train_batch_size=8, per_device_eval_batch_size=8, weight_decay=0.01)
trainer = Trainer(model=model, args=training_args, train_dataset=train_dataset, eval_dataset=val_dataset)
trainer.train()
```

```python
predictions=trainer.predict(test_dataset)
```

```python
test_results=test.copy(deep=True)

temp = []
for x in range(len(test_results)):
  i = (np.argmax(predictions.predictions[x]))
  temp.append(i)

test_results['label_pred_RoBERTa'] = temp
test_results["label_pred_RoBERTa"] = test_results["label_pred_RoBERTa"].apply(lambda x:labels[x])

error_rate_sum_zs=len(test_results[test_results["label"]!=test_results["label_pred_RoBERTa"]])/len(test_results)
print(f'Accuracy of classifier pipeline: {round(100*(1-error_rate_sum_zs), 2)} %')
```

Summation was used for IE axis and for a multiclass of all 16 mbti types, however it required a longer execution time (1h vs 3h) and lower accuracy. Summarization + Zero-Shot Classification

```python
from transformers import BartTokenizer, BartForConditionalGeneration, BartConfig
model_bart = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn').to('cuda') # Run on the GPU
tokenizer_bart = BartTokenizer.from_pretrained('facebook/bart-large-cnn')

def summarize(text):
    inputs = tokenizer_bart([text], truncation=True, max_length=1024, return_tensors='pt').to('cuda')
    summary_ids = model_bart.generate(inputs['input_ids'], num_beams=4, max_length=50, early_stopping=True)
    return([tokenizer_bart.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids][0])

test_results['label_pred_sum_zs']=test_results['text'].apply(lambda x:classifier(summarize(x), candidate_labels=labels)['labels'][0])
error_rate_sum_zs=len(test_results[test_results["label"]!=test_results["label_pred_sum_zs"]])/len(test_results)
print(f'Accuracy of the Summmarization+Zero-Shot classifier pipeline: {round(100*(1-error_rate_sum_zs), 2)} %')
```

## GloVe Word Embedding Config

```python
wordcout = data
wordcout['wordcount'] = data['text'].apply(lambda x: len(str(x).split(' ')))
sequence_length = wordcout.wordcount.max() # max length

print(sequence_length)
```

```python
import urllib.request
import zipfile
import os

glovefile = 'glove.twitter.27B.100d.txt'

# download file if not in file directory
if not os.path.isfile(glovefile):
    urllib.request.urlretrieve('http://nlp.stanford.edu/data/glove.twitter.27B.zip',
                              'glove.twitter.27B.zip')
    with zipfile.ZipFile('glove.twitter.27B.zip', 'r') as zip_ref:
        zip_ref.extractall('')

    print('Download complete!')
    os.remove('glove.twitter.27B.zip')

# creating embeddings index
embeddings_index = {}

f = open(glovefile)
for line in f:
    values = line.split()
    word = values[0]
    vector = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = vector

f.close()

print(len(embeddings_index)) # num of word vectors found
```

```python
max_features = 54000
tokenizer = Tokenizer(num_words=max_features, split=' ', oov_token='<unw>', filters=' ')
tokenizer.fit_on_texts(data['text'].values)

X = tokenizer.texts_to_sequences(data['text'].values)
word_index = tokenizer.word_index
print(len(word_index)) # num of unique tokens
```

```python
num_words = min(max_features, len(word_index)) + 1
print(num_words)
```

```python
embedding_dim = 100
embedding_matrix = np.zeros((num_words, embedding_dim)) # fill with zeros

for word, i in word_index.items():
    if i > max_features:
        continue
    
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
    else:
        embedding_matrix[i] = np.random.randn(embedding_dim) # assign random vector
```
